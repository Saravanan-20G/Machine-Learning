```python

from langchain_community.document_loaders import PyPDFLoader


file_path = r"D:\Saran\RAG\Prompt Engineering and properly use ChatGPT _ LLMs _ Prompt.pdf"

loader = PyPDFLoader(file_path)
documents = loader.load()

print(f"Loaded {len(documents)} pages from the PDF.")
```

```python
from langchain_core.documents import Document
```

```python
def clean_text(documents):
    documents = documents.replace("\n", " ")
    documents = " ".join(documents.split())  # remove extra spaces
    return documents

docs = [
    Document(
        page_content=clean_text(doc.page_content),
        metadata=doc.metadata
    )
    for doc in docs
]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter= RecursiveCharacterTextSplitter(
    separators=["\n●", "\n\n", "\n", ".", " "],
    chunk_size=300,
    chunk_overlap=50
)
chunks=text_splitter.split_documents(documents)
print(chunks)
print(len(chunks))
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('all-MiniLM-L6-v2')

texts = [doc.page_content for doc in chunks]
embeddings=model.encode(texts)


print(embeddings)
```

```python

from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings


embeddings=HuggingFaceEmbeddings()
texts = [doc.page_content for doc in chunks]
db=FAISS.from_texts(texts,embeddings)
query='what is Prompt  Engineering '

docs=db.similarity_search(query)
print(docs)
```

```python
retriever=db.as_retriever()
query='what is Instruction Prompting '
docs = retriever.invoke(query) 

print(docs)
```
