Plot histograms for all numeric columns

```python
sns.histplot(df)
plt.show()
```

```python
df.hist(figsize=(12, 10), bins=20, color='skyblue')
plt.tight_layout()
plt.show()
```

Box Plots to Detect Outliers

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df[['Beats Per Minute (BPM)', 'Energy', 'Danceability', 'Loudness (dB)', 'Popularity']])
plt.title('Boxplot for Numeric Columns')
plt.xticks(rotation=45)
plt.show()
```

Top Genres Analysis

```python
plt.figure(figsize=(10,8))
sns.countplot(y='Top Genre',data=df,order=df['Top Genre'].value_counts().index,palette='viridis')
plt.title('Top Genres Count')
plt.show()
```

```python
sns.countplot(df['Top Genre'])
plt.show()
```

Relationship Between Popularity and Energy

```python
plt.figure(figsize=(20,15))
sns.scatterplot(x='Energy',y='Popularity',data=df,hue='Top Genre',palette='coolwarm')
plt.title('Popularity vs Energy by Genre')
plt.show()
```

 Yearly Trends of Popularity and BPM

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Year',y='Popularity',data=df,label='Popularity', marker='o')
sns.lineplot(x='Year', y='Beats Per Minute (BPM)', data=df, label='BPM', marker='o')
plt.title('Trends of Popularity and BPM Over Years')
plt.legend()
plt.show()
```

Danceability vs Loudness

```python
# Joint plot between Danceability and Loudness
sns.jointplot(x='Danceability', y='Loudness (dB)', data=df, kind='hex', gridsize=30, cmap='Blues')
plt.suptitle('Danceability vs Loudness', y=1.02)
plt.show()
```

Valence and Acousticness Relationship

```python
# Plot relationship between Valence and Acousticness
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Valence', y='Acousticness', data=df, hue='Top Genre', palette='Spectral')
plt.title('Valence vs Acousticness by Genre')
plt.show()
```
