```python
# Example: Create a feature based on distance categories
df['Distance Category'] = pd.cut(df['Distance to the nearest MRT station'], bins=[0, 500, 2000, 4000, 6000], labels=['Very Near', 'Near', 'Far', 'Very Far'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Selecting features and target variable
features = ['Distance to the nearest MRT station', 'Number of convenience stores', 'Latitude', 'Longitude']
target = 'House price of unit area'
```

```python
x=df[features]
y=df[target]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LinearRegression()

model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_train)

print(y_pred)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label='Actual', linestyle='-', marker='o')
plt.plot(y_pred, label='Predicted', linestyle='--', marker='x')
plt.xlabel('Index')
plt.ylabel('House Price')
plt.title('Actual vs. Predicted House Prices')
plt.legend()
plt.show()
```
