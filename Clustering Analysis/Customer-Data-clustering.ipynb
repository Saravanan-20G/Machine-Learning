```python
# Identifying outliers in 'BALANCE' using the IQR method
Q1 = df['BALANCE'].quantile(0.25)
Q3 = df['BALANCE'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers as points beyond 1.5 * IQR from the quartiles
outliers = df[(df['BALANCE'] < (Q1 - 1.5 * IQR)) | (df['BALANCE'] > (Q3 + 1.5 * IQR))]

print(f"Number of outliers in 'BALANCE': {len(outliers)}")
```
