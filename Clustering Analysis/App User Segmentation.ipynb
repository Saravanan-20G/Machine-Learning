Distribution of Numeric Columns

```python
numeric_columns=['Average Screen Time', 'Average Spent on App (INR)', 'Ratings', 'Last Visited Minutes', 'New Password Request']
```

```python
plt.figure(figsize=(15,10))
for i,col in enumerate(numeric_columns,1):
    plt.subplot(2,3,i)
    sns.histplot(df[col],kde=True,bins=50)
    plt.title(f'Distribution of{col}')
plt.tight_layout()
plt.show()
```

Boxplots to check for outliers

```python
plt.figure(figsize=(15,10))
for i,col in enumerate(numeric_columns,1):
    plt.subplot(2,3,i)
    sns.boxplot(df[col])
    plt.title(f'BoxPlot of{col}')
plt.tight_layout()
plt.show()
```

Distribution of Categorical Columns

```python
plt.figure(figsize=(10,6))
sns.countplot(df['Status'])
plt.title('Count of status(Installed vs Uninstalled)')
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.countplot(df['Left Review'])
plt.title('Count of Left Review (1 = Yes, 0 = No)')
plt.show()
```

Compare users by Status

```python
installed = df[df['Status'] == 'Installed']
uninstalled = df[df['Status'] == 'Uninstalled']

plt.figure(figsize=(12,5))
sns.kdeplot(installed['Average Screen Time'], label='Installed', shade=True)
sns.kdeplot(uninstalled['Average Screen Time'], label='Uninstalled', shade=True)
plt.title('Screen Time: Installed vs Uninstalled')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.kdeplot(installed['Average Spent on App (INR)'], label='Installed', shade=True)
sns.kdeplot(uninstalled['Average Spent on App (INR)'], label='Uninstalled', shade=True)
plt.title('Spent on App: Installed vs Uninstalled')
plt.legend()
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(10,6))
corr = df[numeric_columns].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

App User Segmentation to Find Retained and Lost Users

```python
clustering_data = df[["Average Screen Time", "Left Review", 
                        "Ratings", "Last Visited Minutes", 
                        "Average Spent on App (INR)", 
                        "New Password Request"]]

from sklearn.preprocessing import MinMaxScaler
for i in clustering_data.columns:
    MinMaxScaler(i)
    
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(clustering_data)
df["Segments"] = clusters

print(df.head(10))
```

```python
print(df['Segments'].value_counts())
```

```python
df['Segments']=df['Segments'].map({0:'Retained',1:'Churn',2:'Needs Attention'})
```
